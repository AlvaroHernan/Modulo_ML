•	Aplique los hiperparámetros para los tres algoritmos.

•	Finalmente, compare los resultados obtenidos y argumente qué algoritmo es el mejor y por qué.


```python
# Librerías
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
import pandas as pd

# Selección de variables más importantes
X = data[['ph', 'Solids', 'Sulfate', 'Chloramines', 'Hardness']]
y = data['Potability']

# División 70% entrenamiento, 30% prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier

# Modelos
model_nb = GaussianNB()
model_dt = DecisionTreeClassifier(random_state=42)
model_gbc = GradientBoostingClassifier(random_state=42)

# Entrenamiento
model_nb.fit(X_train, y_train)
model_dt.fit(X_train, y_train)
model_gbc.fit(X_train, y_train)
```

```python
# Predicciones
pred_nb = model_nb.predict(X_test)
pred_dt = model_dt.predict(X_test)
pred_gbc = model_gbc.predict(X_test)

# Función para evaluar modelos
def evaluar_modelo(nombre, y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    print(f"🔹 {nombre}")
    print(f"Precision: {precision:.3f} | Recall: {recall:.3f} | F1-score: {f1:.3f}\n")
    return [nombre, precision, recall, f1]

# Evaluar
resultados = []
resultados.append(evaluar_modelo("Naive Bayes", y_test, pred_nb))
resultados.append(evaluar_modelo("Árbol de Decisión", y_test, pred_dt))
resultados.append(evaluar_modelo("Gradient Boosting", y_test, pred_gbc))

# Tabla comparativa
df_resultados = pd.DataFrame(resultados, columns=["Modelo", "Precision", "Recall", "F1-score"])
df_resultados
```

```python
from sklearn.model_selection import GridSearchCV

# Árbol de Decisión
param_dt = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}
grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_dt, cv=5)
grid_dt.fit(X_train, y_train)

# Gradient Boosting
param_gbc = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7]
}
grid_gbc = GridSearchCV(GradientBoostingClassifier(random_state=42), param_gbc, cv=5)
grid_gbc.fit(X_train, y_train)
```

```python
# Predicciones optimizadas
pred_dt_best = grid_dt.best_estimator_.predict(X_test)
pred_gbc_best = grid_gbc.best_estimator_.predict(X_test)

# Evaluar modelos ajustados
resultados_opt = []
resultados_opt.append(evaluar_modelo("Naive Bayes (base)", y_test, pred_nb))
resultados_opt.append(evaluar_modelo("Árbol de Decisión (ajustado)", y_test, pred_dt_best))
resultados_opt.append(evaluar_modelo("Gradient Boosting (ajustado)", y_test, pred_gbc_best))

# Tabla comparativa final
df_final = pd.DataFrame(resultados_opt, columns=["Modelo", "Precision", "Recall", "F1-score"])
df_final
```

Tras aplicar la técnica de mejores características y entrenar los tres modelos de clasificación, se observa que el algoritmo Gradient Boosting Classifier (GBC) obtiene los mejores resultados globales, alcanzando una precisión de 0.73, recall de 0.42 y un F1-score de 0.53.

Estos valores superan significativamente a los obtenidos por Naive Bayes y Árbol de Decisión, lo que demuestra que el enfoque de ensamble de múltiples árboles permite capturar relaciones no lineales entre las variables físico-químicas (pH, sólidos, sulfatos, cloraminas, dureza) y la potabilidad del agua.

En contraste, Naive Bayes presenta un rendimiento limitado debido a su suposición de independencia entre variables, y el Árbol de Decisión muestra indicios de sobreajuste.

Por tanto, se concluye que el Gradient Boosting es el modelo más adecuado para este tipo de datos, al ofrecer el mejor balance entre precisión y sensibilidad en la predicción de la potabilidad del agua.
