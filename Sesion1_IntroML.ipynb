## **4. Limpieza de datos Numéricos y Categoricos**

```python
dat={
    'Country':['Costa Rica','India','Brazil','Colombia','Portugal'],
    'Capital':['San José','New Delhi','Brasilia','Bogotá','Lisboa'],
    'Population':[4500000,150000,80317105,70317105,np.nan],
    'Language':['Spanish','Indi',np.nan,'Spanish','Portugues'],
     'States':[30,np.nan,24,38,2]}

df = pd.DataFrame(dat)
df
```

```python
df.isna().sum()   #Conteo de datos nulos
#Si los datos NaN > 25% de df se debe limpiar
```

```python
#df.dropna()   #Elimina filas con datos nulos cuando los datos no son representativos, último por hacer
# df.info()  Ver columnas y verificar datos nulos
#Cuando son MCAR

df['Population'].fillna(0,inplace=True)
df['Language'].fillna('Desconocido',inplace=True)
df['States'].fillna(0,inplace=True)
df
```

```python
# Datos MNAR
df2=df.copy()
df2
```

```python
#MNAR - MAR
df2
df2['Population'].plot(kind='kde') #Normalidad para saber si una variable puede ser predecida o no 95% de confianza y 5% de margen de error permisible
```

```python
df2['Population'].fillna(df2['Population'].mean(),inplace=True)
df2
```

```python
df2
df2['States'].plot(kind='kde',color='red') #Normalidad
```

```python
df2['States'].fillna(df2['States'].median(),inplace=True)
df2
```

```python
df2['Language'].fillna(df2['Language'].mode()[0],inplace=True)
df2
```
